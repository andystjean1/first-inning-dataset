```python
import pandas as pd
import numpy as np
from os import listdir

import matplotlib.pyplot as plt
```

```python
csv_filenames = listdir("./data/csv_data")

master_df = pd.DataFrame()

for filename in csv_filenames:
    filepath = "/Users/homefolder/Documents/Code/Python/retrosheet_data/data/csv_data/{}".format(filename)
    team_df = pd.read_csv(filepath)
    
    if(master_df.empty):
        master_df = team_df
    else:
        master_df = master_df.append(team_df, ignore_index=True)

master_df.info()
```

```python

temp = master_df.dropna(axis=0)
temp.info()
```

```python
#grab the rows that have full stats
games_df = master_df[master_df["away_WHIP"] != -1]
games_df = games_df.dropna(axis=0)
#label the games where there was a score in the first inning
games_df["first_inning_score"] = np.where(games_df['first_inning_total'] > 0, 1, 0)
#drop the rows with unkown wind direction
games_df["first_inning_total"].hist(bins=12)
```

```python
#split the data into test and training sets
features = ['temperature', 'wind_speed', 'first_home_ba', 'first_home_obp', 'first_home_slg',
       'first_home_ops', 'second_home_ba', 'second_home_obp',
       'second_home_slg', 'second_home_ops', 'third_home_ba', 'third_home_obp',
       'third_home_slg', 'third_home_ops', 'first_away_ba', 'first_away_obp',
       'first_away_slg', 'first_away_ops', 'second_away_ba', 'second_away_obp',
       'second_away_slg', 'second_away_ops', 'third_away_ba', 'third_away_obp',
       'third_away_slg', 'third_away_ops', 'home_ERA', 'home_WHIP', 'home_FIP',
       'home_KOP', 'home_BBP', 'away_ERA', 'away_WHIP', 'away_FIP', 'away_KOP',
       'away_BBP']

X = games_df[features]
y = games_df["first_inning_total"]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=69)

print(len(X_train), len(y_train))
print(len(X_test), len(y_test))
```

```python
#make and evaluate a regression model with all the features
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

#fit and predict
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

#print the metrics
print("Mean squared error: ", mean_squared_error(y_test, y_pred))
print("R Squared score: ", r2_score(y_test, y_pred))
print("coefficients")
print(model.coef_)
```

```python
#make and evaluate a model for each indiviudal feature
for feature in features:
    feat_train = X_train[[feature]]
    feat_test = X_test[[feature]]
    
    model = LinearRegression()
    model.fit(feat_train, y_train)
    y_pred = model.predict(feat_test)
    
    print("Scores for ", feature)
    print("MSE: ", mean_squared_error(y_test, y_pred))
    print("R-squared: ", r2_score(y_test, y_pred))
    print('-'*20)
```
